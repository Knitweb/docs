## 🧪 ChemField Demo

```python
from chemfield import ChemFieldGraph

# Initialize ChemField knowledge graph
chem = ChemFieldGraph()

print("📊 ChemField Statistics:")
print(f"  • Total Nodes: {chem.total_nodes}")
print(f"  • Total Edges: {chem.total_edges}")
print(f"  • Elements: {chem.element_count}")
print(f"  • Compounds: {chem.compound_count}")
print(f"  • Reactions: {chem.reaction_count}")
print(f"  • Properties: {chem.property_count}")
print(f"  • Applications: {chem.application_count}")
print(f"  • Safety Classifications: {chem.safety_count}")
print(f"\n🔬 Knowledge Domains:")
for field in chem.fields:
    print(f"  • {field.title()}")
```

## 🔄 Cross-Field Integration

```python
# Demonstrate cross-field connections
import json

print("🌉 Cross-Field Bridges (ChemField ↔ FinField):")
print()

cross_field_examples = [
    {
        "chemistry": "Vanadium (V)",
        "finance": "VRFB Battery Energy Storage",
        "relation": "component_of",
        "strength": 4
    },
    {
        "chemistry": "Lithium (Li)",
        "finance": "EV Battery Supply Chains",
        "relation": "material_for",
        "strength": 5
    },
    {
        "chemistry": "Sulfuric Acid (H₂SO₄)",
        "finance": "Industrial Production Indices",
        "relation": "input_for",
        "strength": 4
    },
    {
        "chemistry": "Rare Earth Elements (REE)",
        "finance": "Tech Sector Valuations",
        "relation": "component_of",
        "strength": 4
    },
    {
        "chemistry": "Copper (Cu)",
        "finance": "Inflation Hedging",
        "relation": "hedges_against",
        "strength": 3
    }
]

for i, bridge in enumerate(cross_field_examples, 1):
    print(f"{i}. {bridge['chemistry']}")
    print(f"   ↓ [{bridge['relation']}]")
    print(f"   → {bridge['finance']} (strength: {bridge['strength']}/5)")
    print()
```

## 📊 Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np

# ChemField composition
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# ChemField pie chart
chem_data = [
    chem.element_count,
    chem.compound_count,
    chem.reaction_count,
    chem.property_count,
    chem.application_count,
    chem.safety_count
]
chem_labels = ['Elements', 'Compounds', 'Reactions', 'Properties', 'Applications', 'Safety']
colors1 = ['#3399ff', '#33cc33', '#ffcc33', '#cc33ff', '#ff3333', '#ff9933']

ax1.pie(chem_data, labels=chem_labels, autopct='%1.0f%%', colors=colors1, startangle=90)
ax1.set_title('ChemField Composition (308 nodes)', fontsize=12, fontweight='bold')

# FinField bar chart
fin_categories = ['Sectors', 'Industries', 'Companies', 'Metrics', 'Factors', 'Indices']
fin_counts = [11, 51, 25, 15, 14, 14]
colors2 = ['#3399ff', '#33cc33', '#ffcc33', '#cc33ff', '#ff3333', '#ff9933']

ax2.barh(fin_categories, fin_counts, color=colors2)
ax2.set_xlabel('Count', fontsize=10)
ax2.set_title('FinField Entity Distribution (177 nodes)', fontsize=12, fontweight='bold')
ax2.grid(axis='x', alpha=0.3)

for i, v in enumerate(fin_counts):
    ax2.text(v + 1, i, str(v), va='center')

plt.tight_layout()
plt.show()

print("✅ Visualization complete")
```

## 📚 Query the Full Knowledge Graph

```python
# Download full merged graph from 5mart.ml
import requests
import json

print("⏳ Downloading full knowledge graph...")
try:
    graph_url = "https://5mart.ml/cross-field-graph/data.json"
    graph_response = requests.get(graph_url, timeout=10)
    
    if graph_response.status_code == 200:
        graph = graph_response.json()
        
        print("✅ Graph loaded successfully")
        print()
        print(f"📊 Graph Statistics:")
        print(f"  • Nodes: {graph['stats']['total_nodes']}")
        print(f"  • Edges: {graph['stats']['total_edges']}")
        print(f"  • Cross-Field: {graph['stats']['cross_field_edges']}")
        print()
        print(f"🔬 Nodes by Field:")
        for field, count in graph['stats']['nodes_by_field'].items():
            print(f"  • {field.title()}: {count}")
        print()
        print(f"📄 Sample Nodes (first 5):")
        for node in graph['nodes'][:5]:
            print(f"  • {node['id']} ({node['type']})")
    else:
        print(f"⏳ Graph service status: {graph_response.status_code}")
except Exception as e:
    print(f"📡 Check network connection: {str(e)[:50]}")
```
